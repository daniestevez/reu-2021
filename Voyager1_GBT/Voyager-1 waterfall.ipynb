```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
nfft = 256
nfftsb = 512
carrier = np.fromfile('carrier_waterfall.f32', 'float32').reshape((-1, nfft))
sideband = np.fromfile('sideband_waterfall.f32', 'float32').reshape((-1, nfftsb))
```

```python
plt.hist(10*np.log10(carrier).ravel(), bins=100);
```

```python
plt.hist(10*np.log10(sideband).ravel(), bins=100);
```

```python
plt.imshow(10*np.log10(carrier), aspect='auto', vmin=-20, vmax=35)
plt.colorbar();
```

```python
plt.imshow(10*np.log10(sideband), aspect='auto', vmin=10, vmax=35)
plt.colorbar();
```

```python
plt.plot(10*np.log10(np.average(sideband, axis=0)))
plt.ylim((20, 30));
```
